```python
import pandas as pd

# Define the file path to the downloaded dataset
file_path = 'yob2022.txt'

# Load the dataset into a DataFrame
df = pd.read_csv(file_path, names=['Name', 'Gender', 'Count'])

# Display the first few rows of the dataset
print(df.head())

# Basic statistics
print("\nBasic Statistics:")
print(f"Total number of names: {len(df)}")
print(f"Total count of babies: {df['Count'].sum()}")

# Top 10 most popular names
top_10_names = df.groupby('Name')['Count'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Most Popular Names:")
print(top_10_names)

# Visualization: Top 10 names
import matplotlib.pyplot as plt

top_10_names.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Most Popular Baby Names in 2022')
plt.xlabel('Name')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import requests
from io import StringIO

# URL for the SSA baby names dataset for the year 2022
url = 'https://www.ssa.gov/oact/babynames/names.zip'

# Download the dataset
response = requests.get(url)

# Save the content to a file
with open('names.zip', 'wb') as f:
    f.write(response.content)

# Unzip the file and load the data
import zipfile

with zipfile.ZipFile('names.zip', 'r') as zip_ref:
    zip_ref.extractall('names_data')

# Define the file path to the extracted dataset for the year 2022
file_path = 'names_data/yob2022.txt'

# Load the dataset into a DataFrame
df = pd.read_csv(file_path, names=['Name', 'Gender', 'Count'])

# Display the first few rows of the dataset
print(df.head())

# Basic statistics
print("\nBasic Statistics:")
print(f"Total number of names: {len(df)}")
print(f"Total count of babies: {df['Count'].sum()}")

# Top 10 most popular names
top_10_names = df.groupby('Name')['Count'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Most Popular Names:")
print(top_10_names)

# Visualization: Top 10 names
import matplotlib.pyplot as plt

top_10_names.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Most Popular Baby Names in 2022')
plt.xlabel('Name')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import requests
from io import StringIO
import zipfile

# URL for the SSA baby names dataset for the year 2022
url = 'https://www.ssa.gov/oact/babynames/names.zip'

# Download the dataset
response = requests.get(url)

# Save the content to a file
with open('names.zip', 'wb') as f:
    f.write(response.content)

# Unzip the file and load the data
with zipfile.ZipFile('names.zip', 'r') as zip_ref:
    zip_ref.extractall('names_data')

# Define the file path to the extracted dataset for the year 2022
file_path = 'names_data/yob2022.txt'

# Load the dataset into a DataFrame
df = pd.read_csv(file_path, names=['Name', 'Gender', 'Count'])

# Display the first few rows of the dataset
print(df.head())

# Check for missing data
missing_data = df.isnull().sum()
print("\nMissing Data Counts:")
print(missing_data)

# Check for missing values in the DataFrame
print("\nTotal Missing Values:")
print(df.isnull().sum().sum())
```
